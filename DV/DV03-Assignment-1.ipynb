Using the cars data set, first, I will create a parallel coordinates plot using the non-normalized data, and after that using the normalized data. But before all, I need to load the CSV file in pandas.

```python
import pandas as pd
# The data file is quite nasty with several different delimeters that read_csv cannot handle very well
names=['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','car name','j','k','l','m','n']
cars = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data', delimiter=r"\s+", names=names, header=None, engine='python')
# Create a subset of dataset with all useful features
cars = cars.iloc[:,[0,1,2,4,5,6,7]]

# Create a normalized dataset
cars_norm = (cars - cars.mean()) / (cars.max() - cars.min())
```

```python
import numpy as np
from matplotlib import pyplot as plt
from pandas.tools.plotting import parallel_coordinates
%matplotlib inline 

fig = plt.figure(figsize=[15,6])  #creates the fgure
ax = parallel_coordinates(cars_norm,'origin')
ax.set_ylabel('width/length [cm]');
```

### What conclusions can you make from the relation between weight and acceleration? 
Seeing the parallel coordinates plot, I can conclude that as the weight decreases the acceleration increases. We can also clearly see that cars with a more recent model year tend to have a lower weight.

```python
cars_norm = cars_norm[cars['model year']>=80]

fig = plt.subplots(1,2,figsize=[15,6])
ax = parallel_coordinates(cars_norm,'model year')
```

### What conclusions can you draw from cars with model years 80-82?
From the above diagrams we can conclude that cars between '80-'82 started to lower their weight and displacement. Furthermore the acceleration started to increase gradually.

```python
from pandas.tools.plotting import scatter_matrix
%matplotlib inline
scatter_matrix(cars_norm, alpha=1,figsize=(12,12), diagonal='kde', c=['#aaaaaa','#333333','#000000'], s=100, linewidth=0)
```

From the above scatter matrix I can see that through the years cars tend to be more efficient.
